# DATA UNDERSTANDING
- Menggunakan dataset Final.csv. Source : Global Internet users (Internet Users Country-wise 1980-2020 - ASHISHRAUT64) : https://www.kaggle.com/datasets/ashishraut64/internet-users

- Dataset yang digunakan berisi informasi mengenai : 
    - Entity (negara) : Berisi nama negara dan wilayah.
    - Code (kode negara) : Informasi tentang kode negara dan kode mana yang memiliki nilai 'Wilayah', itu menunjukkan pembagian dengan mengelompokkan berbagai negara.
    - Year (tahun) : Tahun 1980-2020
    - Cellular Subscription (jumlah pelanggan layanan seluler) : Langganan ponsel per 100 orang. 
    - Internet Users (%) (persentase pengguna internet) : Pangsa populasi yang mengakses internet untuk semua negara di dunia.
    - No. of Internet Users (jumlah pengguna internet) : Jumlah orang yang menggunakan Internet di setiap negara.
    - Broadband Subscription (jumlah pelanggan internet broadband) : Jumlah langganan broadband tetap per 100 orang. Ini mengacu pada langganan tetap untuk akses berkecepatan tinggi ke Internet publik (koneksi TCP/IP), dengan kecepatan downstream sama dengan, atau lebih besar dari, 256 kbit/dtk.



- Dataset ini digunakan untuk melakukan estimasi persentase populasi pengguna internet di seluruh dunia, sehingga akan fokus pada kolom Internet Users (%) (Persentase pengguna internet).

- Mengidentifikasi variable predictor dan variable response
    - X (variable bebas/predictor) = Cellular Subscription (jumlah pelanggan layanan seluler), No. of Internet Users (jumlah pengguna internet), Broadband Subscription (jumlah pelanggan internet broadband), dan Year (tahun).
    - Y (variable tak bebas/response) = Internet Users (%) (persentase pengguna internet).

# Pemahaman Data dan Visualisasi Data

### Visualisasi Data

```python
df = df.select_dtypes(include=[float, int])

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

### Jumlah pengguna internet berdasarkan negara

```python
jumlah_pengguna_internet = df.groupby('Code').sum()[['No. of Internet Users']].sort_values(by='No. of Internet Users',ascending=True).reset_index()
jumlah_pengguna_internet = jumlah_pengguna_internet.rename(columns={'No. of Internet Users':'Jumlah pengguna internet'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=jumlah_pengguna_internet['Code'], y=jumlah_pengguna_internet['Jumlah pengguna internet'], color='royalblue')
plt.xticks(rotation=60)
```

### Jumlah pengguna internet berdasarkan tahun

```python
jumlah_pengguna_internet_tahun = df.groupby('Year').sum()[['No. of Internet Users']].reset_index()
jumlah_pengguna_internet_tahun = jumlah_pengguna_internet_tahun.rename(columns={'No. of Internet Users':'Jumlah pengguna internet'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=jumlah_pengguna_internet_tahun['Year'], y=jumlah_pengguna_internet_tahun['Jumlah pengguna internet'], color='royalblue')
plt.xticks(rotation=60)
```

# MODELLING DATA
Model yang digunakan untuk menganalisis data yaitu Regresi Linier.

### Seleksi fitur

```python
features = ['Cellular Subscription', 'No. of Internet Users', 'Broadband Subscription', 'Year']
x = df[features]
y = df['Internet Users(%)']
x.shape, y.shape
```

### Split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

### Membuat Model Regresi Linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
for feature in df:
    if feature == 'Internet Users(%)':
        continue
    
    if pd.to_numeric(df[feature], errors='coerce').notnull().all():
        sns.regplot(y=df['Internet Users(%)'],x=df[feature].astype(float))
        plt.show()
```

# EVALUATION DATA

```python
score = lr.score(X_test, y_test)
print('Akurasi model regresi linier = ', score)
```

Didapat milai akurasi model regresi linier sebesar __0.729288408996358__, yang berarti bahwa model dapat menjelaskan sekitar __72,93%__ variasi dalam data. Dalam konteks penelitian ini, hal ini berarti bahwa model regresi linier dapat memperkirakan dengan akurasi yang cukup tinggi persentase populasi pengguna internet di seluruh dunia berdasarkan faktor-faktor yang terkait, seperti jumlah pelanggan layanan seluler, jumlah pengguna internet, jumlah pelanggan internet broadband, tahun dan persentase pengguna internet. Meskipun demikian, masih ada sekitar __27,07%__ variasi dalam data yang tidak dapat dijelaskan oleh model ini, dan faktor-faktor lain mungkin perlu dipertimbangkan untuk memperkirakan persentase pengguna internet di masa depan dengan lebih akurat. Oleh karena itu, perlu dilakukan penelitian lebih lanjut untuk memperbaiki dan meningkatkan model yang digunakan.
